```python
from tensorflow.keras.models import load_model

# Load the model from the file
model_load_path = '/content/drive/My Drive/DeepLearning/my_best_using_30k_images_model.keras'
model = load_model(model_load_path)

print("Model loaded successfully")
```

```python
import pickle
import numpy as np
from keras.preprocessing.image import img_to_array, load_img
from PIL import Image
import io

# Update this path to your test data pickle file location
test_pickle_file_path = '/content/drive/My Drive/DeepLearning/test_images_part4a.pickle'

with open(test_pickle_file_path, 'rb') as f:
    test_images_with_names = pickle.load(f)

# Convert list of tuples into a dictionary for easier access
test_images_dict = {name: img_bytes for name, img_bytes in test_images_with_names}

print("Number of test images:")
print(len(test_images_dict))
```

```python
def preprocess_images(image_dict, target_size=(224, 224)):
    processed_images = []
    image_ids = []
    for image_id, image_bytes in image_dict.items():
        image = Image.open(io.BytesIO(image_bytes))
        image = image.resize(target_size)
        image = img_to_array(image)
        image /= 255.0  # Normalize to [0, 1]
        processed_images.append(image)
        image_ids.append(image_id.replace('.tif', ''))
    return np.array(processed_images), image_ids

X_test, test_ids = preprocess_images(test_images_dict)


!free -h

test_images_dict = None

gc.collect()
```

```python
!free -h
```

```python
predictions = model.predict(X_test)
#predicted_labels = (predictions > 0.5).astype(np.int).reshape(-1)
predicted_labels = (predictions > 0.5).astype(int).reshape(-1)


!free -h
```

```python
import pandas as pd

# Create a DataFrame with test results
test_results_df = pd.DataFrame({
    'id': test_ids,
    'label': predicted_labels
})

# Specify the output file path
output_file_path = '/content/drive/My Drive/DeepLearning/test_predictions_part4a.csv'

# Save the DataFrame to a CSV file
test_results_df.to_csv(output_file_path, index=False)

print("Test predictions saved to:", output_file_path)

!free -h
```
